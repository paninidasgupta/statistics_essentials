```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
ds = xr.open_dataset("2018.nc")
```

## Example

```python
T = 26
depth    =  ds.depth.values
temp     =  ds.water_temp.sel(time="2018-06-30").values
tt       =  ds.water_temp.sel(time="2018-06-30").values-T
i,j      =  np.where(tt>0.0)[0][-1],np.where(tt<0.0)[0][0]
depth_26 =  depth[i] + (T-temp[i])*(depth[j]-depth[i])/(temp[j]-temp[i])
print("Temperature is %f at depth %f "%(ds.water_temp.sel(time="2018-06-30").values[i],depth[i]))
print("temperature is %f at depth %f "%(ds.water_temp.sel(time="2018-06-30").values[j],depth[j]))
print("Therefore, 26C is at depth %f"%depth_26)
```

## Final function

```python
def d26(profile,depth):
    T        =  26.0
    tt       =  profile-T
    i,j      =  np.where(tt>0.0)[0][-1],np.where(tt<0.0)[0][0]
    depth_26 =  depth[i] + (T-profile[i])*(depth[j]-depth[i])/(profile[j]-profile[i])
    return depth_26
```

```python
d26_ds = xr.apply_ufunc(d26,ds['water_temp'],ds.depth,input_core_dims=[['depth'],['depth']],vectorize=True)
# d26_ds.to_netcdf('d26.nc') ## write into nc file
```
